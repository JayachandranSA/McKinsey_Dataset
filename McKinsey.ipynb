```python
#Concatenating
users=pd.DataFrame({'user_id':[1,2,3],'name':["sharadh", "shahid", "khusalli"]})
users
```

```python
msgs=pd.DataFrame({'user_id':[1,1,3,4],'msgs':['hmm', "acha", "theek hai", "nice"]})
msgs
```

```python
pd.concat([users,msgs])
```

```python
#merge
users.merge(msgs,on='user_id',how="outer")
```

```python
users.merge(msgs,on="user_id",how='left')
```
